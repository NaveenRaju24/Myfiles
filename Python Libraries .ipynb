```python
# 1-Numpy ( for array operations, Mathematical and logical functions, Linear algebra, statistics etc...)

# !pip install numpy            # to install it(but here in jupiter notebook this is not required) # 
import numpy as np            # importing the library 

ar1=np.array([10,20,30])      # creating 1D array - one row
print("1D array",ar1) 
ar2=np.array([[10,20],[30,40]])      # creating 2D array - row and columns 
print("2D array/n",ar2)
ar3=np.array([[[10,20],[30,40]]])     # creating 3D array - multiple 2D layers 
print("3D array/n",ar3)

# some inbuilt functions 
print(np.zeros((2, 3)))
print(np.eye(3))
print("Full:/n", np.full((2, 2), 7))

# Properties 
a = np.array([[10, 20, 30], [40, 50, 60]])

print("Array:/n", a)
print("Shape:", a.shape)              # Rows x Columns
print("Dimensions:", a.ndim)          # 1D, 2D, 3D
print("Data Type:", a.dtype)          # int32, float64, etc.

# indexing , slicing 
arr= np.array([10, 20, 30, 40, 50])        # indexing 
print(arr[0])                            # First element → 10 ( it will go like this = 0,1,2,3, and from back side it is ....-3,-2,-1.
print(arr[-1])                           # Last element  → 50
print(arr[-5])
print(arr[1:4])                          # slicing(selecting a range of elements)
print(arr[::2])                          # [10, 30, 50] (step by 2)
matrix = np.array([[1, 2, 3],            # 2D 
                   [4, 5, 6],
                   [7, 8, 9]])

print(matrix[0][1])                      # 2 (row 0, column 1)
print(matrix[1, 2])                      # 6 (row 1, column 2)
print(matrix[0:2, 1:3])                  # 2D slicing 

# Array Operations/Functions in NumPy

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(a + b)                           # [5 7 9]
print(a - b)                           # [-3 -3 -3]
print(a * b)                           # [4 10 18]
print(np.sum(a))                       # Mathemetical # output = 6
print(np.mean(b))                                     # 5.0
print(np.max(a))                                      # 3
print(np.min(b))                                      # 4
print(a > 1)                           # Logical output = [False  True  True]
print(a == 2)                                           # [False  True  False]
a = np.array([[1, 2], [3, 4]])         # Dot Product (Matrix Multiplication)
b = np.array([[5, 6], [7, 8]])         # o/p = [[19 22] [43 50]] 
print(np.dot(a, b))                    # Row 1 of a dot Columns of b: [1, 2] • [5, 7] = (1×5) + (2×7) = 5 + 14 = 19 
                                                                    #[1, 2] • [6, 8] = (1×6) + (2×8) = 6 + 16 = 22
                                       # Row 2 of a dot Columns of b: [3, 4] • [5, 7] = (3×5) + (4×7) = 15 + 28 = 43
                                                                    # [3, 4] • [6, 8] = (3×6) + (4×8) = 18 + 32 = 50
print(np.std(a))                   # Standard deviation
print(np.var(a))                   # Variance
a = np.array([1, 2, 3, 4])           # Cumulative Sums
print(np.cumsum(a))                  # [1 3 6 10]
print(np.cumprod(a))                 # [1 2 6 24]
```

```python
# NumPy Broadcasting (to allow element-wise operations between arrays)

import numpy as np
arr = np.array([1, 2, 3])
result = arr + 10  # Broadcasting scalar 10 to each element
print(result)

mat = np.array([[1, 2, 3],      # 2D 
                [4, 5, 6]])
vec = np.array([1, 0, -1])
result = mat + vec
print(result)

a = np.array([[1, 2, 3],      # dimensions must be equal or one of them must be 1
              [4, 5, 6]])     # shape = (2, 3)
b = np.array([[10, 20]])      # shape = (1, 2)
a + b                         # this will give an error 
```

```python
# Numpy Universal Functions (ufuncs) - element-wise operations provided by NumPy

# Categories of ufuncs:
#Arithmetic – np.add(), np.subtract(), np.multiply(), np.divide()
#Comparison – np.greater(), np.equal(), np.less_equal()
#Trigonometric – np.sin(), np.cos(), np.tan()
#Exponential & Logarithmic – np.exp(), np.log(), np.log10()
#Rounding – np.floor(), np.ceil(), np.round()
#Aggregate – np.sum(), np.mean(), np.std()

arr = np.array([1, 2, 3, 4])

print(np.add(arr, 10))          # Arithmetic # [11 12 13 14]                        
print(np.multiply(arr, 2))      # [2  4  6  8]

print(np.exp(arr))              # Exponential & Log # e^x
print(np.log(arr))              # natural log

print(np.sin(np.pi * arr / 2))  # Trigonometry # sin(π/2), sin(π), ...

print(np.greater(arr, 2))       # Comparison # [False False  True  True]

float_arr = np.array([1.3, 2.7, 3.6]) # Rounding
print(np.floor(float_arr))            # [1. 2. 3.]
print(np.round(float_arr))            # [1. 3. 4.]
```

```python
# Numpy Boolean Masking & Filtering ( way to filter or select elements from a NumPy array based on a condition )

arr = np.array([5, 10, 15, 20])
mask = arr > 10
print(mask)        # [False False  True  True]
filtered = arr[mask]
print(filtered)    # [15 20]         # filtering values with true 

print(arr[arr > 10])   # [15 20]     # we can even write directly like this 

results = arr[(arr>=10) & (arr<=20)]   # Elements between 10 and 20
print(results)

mat = np.array([[1, 2, 3],             # on 2D
                [4, 5, 6]])
print(mat[mat > 3])                    # [4 5 6]

arr = np.array([5, 15, 25, 35])
arr[arr > 20] = 99                     # replacing the eliments with condition
print(arr)                             # [ 5 15 99 99]
```

```python
# Numpy array Reshaping, Transposing, and Flattening Arrays
# ( Change an array's shape without changing its data, Convert between 1D/2D/3D, Transpose rows ↔ columns, Flatten multidimensional arrays into 1D

# reshape()- Changes the shape of an array — must maintain the same number of elements.
arr = np.array([1, 2, 3, 4, 5, 6])
reshaped = arr.reshape((2, 3))         # 2 rows, 3 columns
print(reshaped)

arr = np.array([1, 2, 3, 4, 5, 6])    # automatic calculation if we put -1 in the no. of columns 
print(arr.reshape((3, -1)))           # NumPy calculates columns

# flatten() vs ravel() - Both convert multi-dimensional arrays into 1D
mat = np.array([[1, 2], [3, 4]])
flat1 = mat.flatten()                # Returns a copy
flat2 = mat.ravel()                  # Returns a view (faster)
print(flat1)
print(flat2)

# transpose() or .T = Swap rows and columns
mat = np.array([[1, 2, 3],
                [4, 5, 6]])
print(mat.T)                         

# reshape vs resize ( reshape = Does NOT change the original array, Returns a new view or copy .. resize= Modifies the array in-place)
a = np.array([1, 2, 3, 4])
a.resize((2, 2))                     # Modifies in place
print(a)
b=np.array([1,2,3,4])
c= b.reshape((2,2))
print(c)
print(b)                             # it is not changed 
```

```python
# Numpy Stacking & Splitting Arrays (In real-world tasks like feature engineering u need to Combine arrays vertically or horizontally , split them)

# STAKING Arrays
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
vstacked = np.vstack((a, b))       # vertical stack
print(vstacked)

a = np.array([[1], [2], [3]])
b = np.array([[4], [5], [6]])
hstacked = np.hstack((a, b))       # horizontal
print(hstacked)

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
stacked = np.stack((a, b), axis=0)         # shape (2, 3)
print(stacked)
stacked_axis1 = np.stack((a, b), axis=1)   # shape (3, 2)
print(stacked_axis1)

# SPLITTING Arrays
arr = np.array([[1, 2, 3, 4],                # Split horizontally
                [5, 6, 7, 8]])
hsplits = np.hsplit(arr, 2)
print(hsplits[0])  # [[1 2], [5 6]]

vsplit_arr = np.vsplit(arr, 2)               # vertical 
print(vsplit_arr[0])  # [[1 2 3 4]]
 
arr = np.array([10, 20, 30, 40, 50, 60])     # general
split_arr = np.split(arr, 3)
print(split_arr)  # [array([10, 20]), array([30, 40]), array([50, 60])]
```

```python
# NumPy Linear Algebra ( Linear algebra is the foundation for machine learning, deep learning, and data science )

# Matrix Multiplication (np.dot() or @)
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])
dot1 = np.dot(A, B)
dot2 = A @ B                  # Same as above (modern syntax)
print(dot1)
print(dot2)

# Transpose a Matrix (.T)
print(A.T)

# Determinant (np.linalg.det())
det = np.linalg.det(A)           # 
print(det)                       # -2.0

# Inverse of a Matrix (np.linalg.inv())
inv_A = np.linalg.inv(A)
print(inv_A)

# Identity Matrix (np.eye())
I = np.eye(3)
print(I)

# Eigenvalues & Eigenvectors
vals, vecs = np.linalg.eig(A)
print("Eigenvalues:", vals)
print("Eigenvectors:/n", vecs)
```

```python
# Pandas Grouping and Aggregating ( used to split data into groups, apply a function, and then combine the result )

import pandas as pd 
People= pd.read_csv("C:/Users/navee/Desktop/IDEs/Python/Practice/Practice Datasets/sample_people.csv")
Product= pd.read_excel("C:/Users/navee/Desktop/IDEs/Python/Practice/Practice Datasets/sample_products.xlsx")

# Basic group 
People.groupby('City')['Age'].mean()                      # Average Age per City
People.groupby('City')['Name'].count()                        # Number of people per City
People.groupby('City')['Age'].sum()                           # Total Age per City
People.groupby('City')['Age'].agg(['mean', 'min', 'max'])     # Stats by City
People.groupby(['City', 'Age'])['Name'].count()               # Group by multiple columns
People.groupby('City')['Age'].mean().reset_index()            # Reset index after groupby
People.groupby('City').agg({                                  # Apply custom function .agg
    'Age': 'mean',
    'Name': 'count'
})
```

```python
# Pandas Merging, Joining & Concatenating DataFrames(when you have multiple datasets and need to combine them)

Product1= pd.read_csv("C:/Users/navee/Desktop/IDEs/Python/Practice/Practice Datasets/customers.csv")
Product2= pd.read_csv("C:/Users/navee/Desktop/IDEs/Python/Practice/Practice Datasets/orders.csv")

# Concatenation
pd.concat([Product1, Product2])                   # Vertical stacking (row-wise) - Combine one below the other
pd.concat([Product1, Product2], axis=1)           # Side-by-side

# Merging — SQL-style joins
pd.merge(Product1, Product2, on='CustomerID')                   # Inner join - Only rows with matching ID in both
pd.merge(Product1, Product2, on='CustomerID', how='left')       # Left join (all rows from left, match from right)-so if right side not macthing - NaN will come
pd.merge(Product1, Product2, on='CustomerID', how='right')      # Right join (all rows from right)
pd.merge(Product1, Product2, on='CustomerID', how='outer')      # Outer join (all rows from both)
pd.merge(Product1, Product2, left_on='CustomerID', right_on='OrderID')  # Merge on different column names 
```

```python
# Matplotlib Plot Types

# Line Plot
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.title("Line Plot")
plt.show()

# Scatter Plot
plt.scatter(x, y, color='red', marker='o')
plt.title("Scatter Plot")
plt.show()

# Bar Chart
names = ['Alice', 'Bob', 'Charlie']
scores = [85, 90, 75]

plt.bar(names, scores)
plt.title("Bar Chart")
plt.ylabel("Scores")
plt.show()

# Horizontal Bar Chart
plt.barh(names, scores, color='green')
plt.title("Horizontal Bar Chart")
plt.xlabel("Scores")
plt.show()

# Histogram
data = [10, 20, 20, 20, 30, 30, 40, 50, 50, 60, 70, 80]

plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Pie Chart
labels = ['Python', 'Java', 'C++']
sizes = [45, 30, 25]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart")
plt.axis('equal')  # Equal aspect ratio
plt.show()

# Boxplot
data = [10, 20, 30, 40, 50, 60, 70, 100]
plt.boxplot(data)
plt.title("Boxplot")
plt.show()
```

```python
# Matplotlib Subplots and Multiple Figures

# plt.subplot() — Quick Grids (Stateful Style)
# Syntax: plt.subplot(nrows, ncols, index)
plt.subplot(1, 2, 1)        # 1 row, 2 columns, 1st plot
plt.plot([1, 2, 3], [1, 4, 9])
plt.title("Left Plot")

plt.subplot(1, 2, 2)        # 2nd plot
plt.plot([1, 2, 3], [1, 2, 3])
plt.title("Right Plot")

plt.tight_layout()
plt.show()

# plt.subplots() — Recommended (Object-Oriented Style)
fig, ax = plt.subplots(1, 2)         # 1 row, 2 columns → 2 axes
ax[0].plot([1, 2, 3], [1, 4, 9])
ax[0].set_title("Left")

ax[1].plot([1, 2, 3], [1, 2, 3])
ax[1].set_title("Right")

plt.tight_layout()
plt.show()

# Multiple Rows & Columns
fig, axs = plt.subplots(2, 2)   # 2x2 grid = 4 plots
axs[0, 0].plot([1, 2], [3, 4])
axs[0, 1].bar([1, 2], [5, 6])
axs[1, 0].scatter([1, 2], [7, 8])
axs[1, 1].hist([1, 2, 2, 3])

plt.tight_layout()
plt.show()
```

```python
# Matplotlib Legends, Grids, and Ticks Customization

# Legends(Used when you have multiple lines or categories)
plt.plot([1, 2, 3], [4, 5, 6], label='Line A')
plt.plot([1, 2, 3], [6, 5, 4], label='Line B')
plt.legend()                             # Show legend
plt.title("Legend Example")
plt.show()

# Grid Lines(Helpful in reading values)
plt.plot([1, 2, 3], [4, 5, 6])
plt.grid(True)                          # Add grid
plt.title("With Grid")
plt.show()

# Customizing Ticks
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

plt.plot(x, y)
plt.xticks([1, 2, 3, 4], ['A', 'B', 'C', 'D'])   # Custom labels
plt.yticks(rotation=45)                         # Rotate ticks
plt.title("Custom Ticks")
plt.show()
```

```python
# Matplotlib Text and Annotations (This is where you label specific data points or add custom messages directly on your plot) 

# Add Text Inside Plot: plt.text()
x = [1, 2, 3]
y = [10, 20, 30]

plt.plot(x, y)
plt.text(2, 20, "Middle Point", fontsize=12, color='red')  # (x, y, text)
plt.title("Using plt.text()")
plt.show()

# Point Annotation: plt.annotate
plt.plot(x, y)

# Annotate the 3rd point
plt.annotate(
    "Highest Point",
    xy=(3, 30),                    # Point to mark
    xytext=(2, 35),                # Text location
    arrowprops=dict(facecolor='black', arrowstyle="->")
)

plt.title("Using annotate()")
plt.show()
```

```python
# Matplotlib Advanced Plots: Error Bars, Filled Areas, and 3D Plots

# Error Bars — plt.errorbar() -- yerr: vertical error, capsize: length of error bar cap, fmt='o': plot points as circles...
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
errors = [0.5, 0.4, 0.6, 0.3]

plt.errorbar(x, y, yerr=errors, fmt='o', capsize=5, color='blue')
plt.title("Error Bars Example")
plt.show()

# Filled Area — plt.fill_between() - Used to shade between two curves or between a curve and baseline
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.fill_between(x, y, alpha=0.3, color='green')  # Fill under the curve
plt.title("Filled Area Plot")
plt.show()

# 3D Plots — mpl_toolkits.mplot3d
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = [1, 2, 3]
y = [4, 5, 6]
z = [7, 8, 9]

ax.scatter(x, y, z)
ax.set_title("3D Scatter Plot")
plt.show()
```

```python
# Matplotlib Pandas Integration 

# Line Plot from DataFrame
import pandas as pd

data = {
    'Year': [2020, 2021, 2022],
    'Sales': [250, 300, 400]
}
df = pd.DataFrame(data)

df.plot(x='Year', y='Sales', kind='line', title="Sales Over Years")
plt.show()

# Bar Chart
df.plot(x='Year', y='Sales', kind='bar', color='skyblue', title="Bar Chart")
plt.show()

# Pie Chart
df = pd.DataFrame({'Category': ['A', 'B', 'C'], 'Values': [40, 30, 30]})
df.set_index('Category').plot(kind='pie', y='Values', autopct='%1.0f%%')
plt.ylabel("")  # Hide y-label
plt.title("Pie Chart from DataFrame")
plt.show()
```

```python
# Seaborn ( same like matplotlib but some extra features ) 

# Bar plot - Matplotlib vs Seaborn
# Matplotlib 
import matplotlib.pyplot as plt

names = ['Alice', 'Bob', 'Charlie']            # You need to provide raw X & Y data, and handle everything manually
scores = [85, 90, 75]

plt.bar(names, scores, color='skyblue')
plt.title("Bar Plot - Matplotlib")
plt.xlabel("Name")
plt.ylabel("Score")
plt.show()

# Seaborn
import seaborn as sns
import pandas as pd

# Create a DataFrame
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'Alice', 'Bob', 'Charlie'],
    'Score': [85, 90, 75, 80, 92, 78]
})

sns.barplot(data=df, x='Name', y='Score')   # Automatically shows mean bars
plt.title("Bar Plot - Seaborn")
plt.show()





```

```python
# Data visualization 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data
data = {
    'Product': ['A', 'B', 'C', 'D'],
    'Sales': [100, 150, 80, 120]
}
Product = pd.DataFrame(data)

# Bar chart
plt.figure(figsize=(6,4))
sns.barplot(x='Product', y='Sales', data=Product)
plt.title("Sales by Product")
plt.ylabel("Units Sold")
plt.xlabel("Product")
plt.show()

```
